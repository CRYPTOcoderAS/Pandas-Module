## DROP NaN VALUES

```python
# DROP NaN rows -> axis=0 default
df.dropna()
```

```python
# DROP NaN columns
df.dropna(axis=1)
```

```python
# drop the row if all has NaN value
df.dropna(how='all')
```

```python
# drop the row in any has NaN -> default how=any
df.dropna(how='any')
```

```python
#thresh -> NOT-NULL values
#dont drop the row having 'n' NOT NULL values
df.dropna(thresh=1)
```

```python
#thresh -> NOT-NULL values
#dont drop the row having 'n' NOT NULL values
df.dropna(thresh=8)
```

```python
#subset 
#drop row while checking the specific subset column
df.dropna(subset=['Name'])
```
