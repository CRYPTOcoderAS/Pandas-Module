## FILL NaN VALUES

```python
import pandas as pd
```

```python
df=pd.read_csv('C:\\Users\\lalit\\Documents\\JUPYTER NOTEBOOK\\Fortune_10.csv',na_values={'Name':'Techdrill'}) 
df
```

```python
# fill zero '0' in the NaN value place
df.fillna(0)
```

```python
# fill specific column with specific value
df.fillna({'Name':'---', 'Expenses':'ZERO'})
```

```python
# ffill -> fills the above cell value (forward fill->above)
# pad -> similar to ffill  (fills previous value)
df.fillna(method='ffill')
# here Mattouch from above cell
```

```python
# bfill -> fills the down cell value (backward fill->down)
df.fillna(method='bfill')
#here Techline from down cell
```

```python
# row fill (axis=0)
df.fillna(method='ffill',axis=0)
```

```python
# column fill (axis=1)
df.fillna(method='ffill',axis=1)
```

```python
# limit -> fills limited int values
df.fillna(0,limit=1)
#here 1 means only one column in each row -> (1 NaN value only)
```

## Pandas merge()

```python
df1=pd.DataFrame({'ID': [1,2,3,4,5],
                 'Class':[9,10,11,12,13]})
df1
```

```python
df2=pd.DataFrame({'ID': [1,2,3,4,5],
                 'Name':[20,21,22,23,24]})
df2
```

```python
pd.merge(df1,df2,on='ID')
```

```python
df2=pd.DataFrame({'ID': [1,2,3,4,6],
                 'Name':[20,21,22,23,24]})
df2
```

```python
pd.merge(df1,df2,on='ID')
# merge and prints only similar IDs
```

```python
#default inner
pd.merge(df1,df2,on='ID',how='inner')
```

```python
# left merge
pd.merge(df1,df2,on='ID',how='left')
```

```python
#outer merge
pd.merge(df1,df2,on='ID',how='outer')
```

```python
# indicator() 
#default inner
pd.merge(df1,df2,on='ID',how='outer',indicator=True)
```

```python
df2=pd.DataFrame({'ID': [6,7,8,9,10],
                 'Name':[20,21,22,23,24]})
df2
```

```python
#having no commmon index
pd.merge(df1,df2,left_index=True, right_index=True)
```

```python
df3=pd.DataFrame({'ID': [1,2,3,4,5],
                 'Name':[20,21,22,23,24]})
df3
```

```python
df4=pd.DataFrame({'ID': [1,2,3,4,5],
                 'Name':[220,221,222,233,224]})
df4
```

```python
# having all same indexes
pd.merge(df3,df4,on='ID')
```

```python
# Change column names
pd.merge(df3,df4,on='ID',suffixes=('_High','_Low'))
```

## Pandas concat()

```python
sr1=pd.Series([0,1,2,3])
sr1
```

```python
sr2=pd.Series([4,5,6,7,8])
sr2
```

```python
#concate same size series
pd.concat([sr1,sr2])
```

```python
sr2=pd.Series([4,5,6,7,9,9,8])
sr2
```

```python
##concate different size series 
pd.concat([sr1,sr2])
```

```python
df01=pd.DataFrame({'ID':[1,2,3,4],
                  'NAME': ['A','B','C','D'],
                  'Class':[5,6,7,8]})
df01
```

```python
df02=pd.DataFrame({'ID':[5,6,7,8],
                  'NAME': ['E','F','G','H'],
                  'Class':[5,6,7,8]})
df02
```

```python
# concate two dataframes
pd.concat([df01,df02])
```

```python
# concate two dataframes column-wise (row same)
pd.concat([df01,df02],axis=1)
```

```python
# concate two dataframes (ignore index)
pd.concat([df01,df02],ignore_index=True)
```

```python
df03=pd.DataFrame({'ID':[1,2,3,4,5,6],
                  'NAME': ['A','B','C','D','E','F'],
                  'Class':[5,6,7,8,9,10]})
df03
```

```python
# join 'outer' default -> union 
pd.concat([df01,df03],axis=1)
```

```python
#join 'inner' default -> intersection
pd.concat([df01,df03],axis=1,join='inner')
```

```python
# keys 
pd.concat([df01,df03],keys=['df01','df03'])
```

```python
pd.concat([df01,df03],axis=1,keys=['FIRST DF','SECOND DF'])
```

## Pandas join()

```python
df01=pd.DataFrame({'ID':[1,2,3,4],
                  'NAME': ['A','B','C','D'],
                  'Class':[5,6,7,8]})
df02=pd.DataFrame({'ID2':[5,6,7,8],
                  'NAME2': ['A','B','C','D'],
                  'Class2':[5,6,7,8]})
display(df01,df02)
```

```python
df01.join(df02)
# must have same index 0,1,2 
```

```python
df02=pd.DataFrame({'ID2':[5,6],
                  'NAME2': ['A','B'],
                  'Class2':[5,6]})
df02
```

```python
df01.join(df02) 
```

```python
df01.join(df02,how='inner')
```

```python
df02=pd.DataFrame({'ID':[5,6,7,8],
                  'NAME': ['A','B','C','D'],
                  'Class02':[5,6,7,8]})
df02
```

```python
# overlapping columns [ID,NAME,CLASS]
df01.join(df02,lsuffix='_1')
#changes name of left df columns having same names
```

## Pandas Pivot Table 

```python
df=pd.read_csv('C:\\Users\\lalit\\Documents\\JUPYTER NOTEBOOK\\Movie-Rating.csv')
df
```

```python
df.pivot_table(index='Genre')
# shows mean value default
```

```python
# to show specific column wise result
df.pivot_table(index='Genre',columns='Year')
```

```python
# aggregate function (sum,mean,standard deviation,min,max,count,etc)
df.pivot_table(index='Genre',columns='Year',aggfunc='count')
```

```python
df.pivot_table(index='Genre',columns='Year',aggfunc='sum')
```

```python
# fill_value -> fills NaN values
df.pivot_table(index='Genre',columns='Year',fill_value='none') 
```

```python
#margins-> average of all rows or columns
df.pivot_table(index='Genre',columns='Year',fill_value='none',margins=True) 
```

```python
#shows the total sum 
df.pivot_table(index='Genre',columns='Year',aggfunc='sum',fill_value='none',margins=True) 
```

## Pandas melt()
